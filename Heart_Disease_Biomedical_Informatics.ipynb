```python
df.groupby('target')['age'].mean()
```

```python
# In this dataset, patients with heart disease have a lower average age than those without disease, suggesting dataset-specific patterns rather than general population risk.
```

```python
df.groupby('target')['age'].describe()
```

```python
# Age distribution shows that patients with heart disease include younger individuals and have a wider age spread, indicating dataset-specific referral or sampling patterns rather than population-level risk.
```

```python
df.groupby('target')['trestbps'].mean()
```

```python
# Patients with heart disease show lower average resting blood pressure, possibly reflecting treatment effects or clinical management rather than true baseline risk.
```

```python
df.groupby('target')['chol'].mean()
```

```python
# Patients with heart disease show lower average cholesterol levels, likely reflecting treatment effects and clinical management rather than underlying baseline risk.
```

```python
import matplotlib.pyplot as plt

# Count of patients with and without heart disease
df['target'].value_counts().plot(kind='bar')

# Labels and title
plt.xlabel("Heart Disease Status (0 = No, 1 = Yes)")
plt.ylabel("Number of Patients")
plt.title("Distribution of Heart Disease in the Dataset")

plt.show()
```

```python
# The bar chart shows a nearly balanced distribution of patients with and without heart disease, supporting reliable comparative analysis.
```
